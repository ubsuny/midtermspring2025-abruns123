```python
import numpy as np
from scipy.linalg import eig
```

```python
m1=15.99
m2=12.011
M = np.diag([m1,m2,m2,m2,m1])
```

```python
k12 = 1.532
k23 = 15.98
k34=k23
k45=k12
```

```python
K = np.array([
    [   k12,    -k12,       0,     0,     0  ],
    [   -k12,  k12 + k23, -k23,    0,     0  ],
    [   0.0,    -k23,   k23+k34,  -k34,   0  ],
    [   0,       0,      -k34,   k34+k45,-k45], 
    [   0,       0,        0,     -k45,   k45]
])
```

```python
omega,v = eig(K,M)
```

```python
print("Eigenvalues : ")
print (omega)

print("Eigenvectors:")
print(v)
```
